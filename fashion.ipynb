```
#import all the required libaray

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```
tf.__version__
```

```
#loading the datasets of fashion with 60000 images and there respective labels 
data = keras.datasets.fashion_mnist
```

```
#spliting the data as training set and testing set and there respective labels
(train_images, train_labels), (test_images, test_labels) = data.load_data()

print("training images..... : {}   training labels....: {}".format(train_images[6],train_labels[6]))
```

```
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
#lets see some sample of datasets
for i in range(10):
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel("Actual Class Name :" + class_names[train_labels[i]])
    plt.show()
```

```
train_images = train_images/255.0
test_images = test_images/255.0
```

```
#creating the model and training the model on the dataset

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(128, activation="relu"),
                          keras.layers.Dense(10, activation="softmax")
                          ])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(train_images, train_labels, epochs=30)

test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```
print("Tested Acc : ", test_acc)
```

```
#lets make some prediction with the model

prediction = model.predict(test_images)
print(class_names[np.argmax(prediction[9])])
```
